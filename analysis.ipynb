This is code for importing the file format as output by the CH intruments analyzer program.

Then at the bottom I actually do the imports and name them `cv_data` (for cyclic voltametry) and `hd_data` (for hydrodynamic cyclic voltametry).

```python
def import_data(filename):
    first = False
    number = None
    results=[]
    with open(filename,'r') as f:
        for line in f.readlines():
            match = re.match("Segment (\d*):",line)
            if match:
                number = int(match.group(1))
                if 1 == number:
                    first = not first
                if first:
                    results.append({"Ep":None,"Eh":None,"data":[]})
            elif number:
                ep = re.match("Ep = (\d*\.\d*)V",line)
                eh = re.match("Eh = (\d*\.\d*)V",line)
                data = re.match("(-*\d*\.\d*), (-*\d*\.\d*e*-*\d*)",line)
                if ep and first:
                    results[number-1]["Ep"] = float(ep.group(1))
                elif eh and first:
                    results[number-1]["Eh"] = float(eh.group(1))
                elif data and not first:
                    results[number-1]["data"].append([float(data.group(1)),float(data.group(2))])
    results = [{"Ep": result["Ep"],"Eh": result["Eh"],"data":np.array(result["data"])} for result in results] 
    return results

data_10mVs = import_data("part ii - 10 mv s-1.txt")
data_50mVs = import_data("part ii - 50 mv s-1.txt")
data_100mVs = import_data("part ii - 100 mv s-1.txt")
data_200mVs = import_data("part ii - 200 mv s-1.txt")
data_1000mVs = import_data("part ii - 1000 mv s-1.txt")
data_300rpm = import_data("part iii - 100 mv s-1 300 rpm.txt")
data_600rpm = import_data("part iii - 100 mv s-1 600 rpm.txt")
data_1200rpm = import_data("part iii - 100 mv s-1 1200 rpm.txt")
data_1500rpm = import_data("part iii - 100 mv s-1 1500 rpm.txt")
data_2000rpm = import_data("part iii - 100 mv s-1 2000 rpm.txt")

cv_data = [data_10mVs,data_50mVs,data_100mVs,data_200mVs,data_1000mVs]
hd_data = [data_300rpm,data_600rpm,data_1200rpm,data_1500rpm,data_2000rpm]
all_data = [data_10mVs,data_50mVs,data_100mVs,data_200mVs,data_1000mVs,data_300rpm,data_600rpm,data_1200rpm,data_1500rpm,data_2000rpm]
```

```python
# calculate number of electrons per molecule
# returns n, ehalf, i_ratio, i_an_peak

def y(a,b,x):
    return a*x + b

def get_values(forward,reverse,flat_forward_index,flat_reverse_index):
    num_samples = reverse.shape[0]
    forward_peak_index = np.argmax(forward[:,1])
    reverse_peak_index = np.argmin(reverse[:,1])
    forward_peak_v,forward_peak_i = forward[forward_peak_index]
    reverse_peak_v,reverse_peak_i = reverse[reverse_peak_index]
    
    # line extrapolation
    a_for,b_for = np.polyfit(forward[0:flat_forward_index,0],forward[0:flat_forward_index,1],1)
    a_rev,b_rev = np.polyfit(reverse[flat_reverse_index:-1,0],reverse[flat_reverse_index:-1,1],1)
    i_for_base = y(a_for,b_for,forward_peak_v)
    i_rev_base = y(a_rev,b_rev,reverse_peak_v)
    i_pa = forward_peak_i - i_for_base
    i_pc = i_rev_base - reverse_peak_i
    return .059/(forward_peak_v-reverse_peak_v),(forward_peak_v+reverse_peak_v)/2., i_pa/i_pc,i_pa

# display cumulative results.
ns = []
ehalfs = []
i_ratios = []
i_an_peaks = []
for data in cv_data:
    sum_n = 0
    sum_ehalf = 0
    sum_iratio = 0
    sum_i_an_peak = 0
    for i in range(0,10,2):
        n, ehalf, i_ratio,i_an_peak = get_values(data[i]['data'],data[i+1]['data'],300,600)
        sum_n += n
        sum_ehalf += ehalf
        sum_iratio += i_ratio
        sum_i_an_peak += i_an_peak
    avg_n = sum_n/5.0
    avg_ehalf = sum_ehalf/5.0
    avg_iratio = sum_iratio/5.0
    avg_i_an_peak = sum_i_an_peak/5.0
    ns.append(avg_n)
    ehalfs.append(avg_ehalf)
    i_ratios.append(avg_iratio)
    i_an_peaks.append(avg_i_an_peak)

print """For each graph, I've printed out the calculated n, E_{1/2} and Ipa/Ipc ratios.
    (Current ratios are calculated by extrapolation of linear parts of graphs as shown
    in lab handout and results are averaged over all five trials per experiment).\n"""
rates = np.array([0.01,0.05,0.1,0.2,1.0])
for i,(n,eh,curr_rat,peak_curr,rate) in enumerate(zip(ns,ehalfs,i_ratios,i_an_peaks,rates)):
    print "Experiment {} ({} mV/s):".format(i+1,rate)
    print "   average n:                    {}".format(n)
    print "   average E_1/2:                {}".format(eh)
    print "   average Ipa/Ipc:              {}".format(curr_rat)
    print "   average peak aniodic current: {}".format(peak_curr)
    print

    
# linear regression of rates and ratios
x_d = rates**(1./2.)
y_d = i_ratios
a_d,b_d = np.polyfit(x_d,y_d,1)
plt.plot(x_d,y_d)
plt.plot(rates,[y(a_d,b_d,rate) for rate in rates])
plt.xlabel("mV/s")
plt.ylabel("Ipa/Ipc")
plt.title("Peak current ratios at differing sweep rates")
plt.show()
'''
#ignore this, this is an incorrect method for calculating diffusion constant.
def diffusion_const(ipf,n,v):
    A = 3.141*((.33)/2)**2
    C = 0.002e-3
    K = 2.69e5
    D = ipf**2/((K*A*C)**2 * n**3 * v)
    return D

for i in range(5):
    print diffusion_const(i_an_peaks[i],ns[i],rates[i])
'''
A = 3.141*((.33)/2)**2
C = 0.002e-3
diff_const = (1.38196e-11*(b_d)**2)/((A)**2*(C)**2*(1)**3)
print "Diffusion constant: {} cm^2/s".format(diff_const)
```
